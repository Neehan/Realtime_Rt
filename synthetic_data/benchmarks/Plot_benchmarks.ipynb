```
import matplotlib.pyplot as plt
import seaborn as sns
from glob import glob
import pandas as pd

%matplotlib inline

sns.set()
```

```
benchmarks = glob('./Realtime_Rt/synthetic_data/benchmarks/*')
benchmarks
```

```
fig, axes = plt.subplots(nrows=len(benchmarks), ncols=3, figsize=(20, 20))

for i, bm in enumerate(benchmarks):
    axs = axes[i, :]
    data = pd.read_csv(bm)
    for j, to_plot in enumerate(['R_t', 'I_t', 'T_t']):
        ax = axs[j]
        y = data[to_plot]
        sns.lineplot(x=range(len(y)), y=y, ax=ax, 
                     label=f"{to_plot}")
        if j == 1:
            ax.set_title(bm.split('/')[-1].split('.')[0], fontsize=16)
        
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import numpy as np

sns.set()

%matplotlib inline
```

```python
found_rt = pd.read_csv('rt/synthetic/bd.csv')
real_rt = pd.read_csv('synthetic_data/new_testing/bd_smooth.csv')
```

```python
scale_const = 1.
offset = 34 - 5.2

ax = sns.lineplot(x=real_rt.R_t.index, y=real_rt.R_t, label='Real $R_t$')
y = found_rt.R_t_mean* scale_const
ax = sns.lineplot(x=found_rt.R_t_mean.index + offset, y=y, ax=ax, label='(Offset) found $\widehat{R_t}$')

# ci = found_rt.R_t_std * 1.96
ax.fill_between(found_rt.R_t_mean.index + offset, found_rt.R_t_low, found_rt.R_t_high, color='b', alpha=.1)

ax.set_ylim(0.5, 3)
ax.set_xlim(18, 115)
ax.set_ylabel('$\widehat{R_t}$')
ax.set_xlabel('Days')
```

```python
ax = sns.lineplot(x=real_rt.I_t.index, y=real_rt.I_t, label='Real $I_t$')
y = found_rt.I_t_mean
ax = sns.lineplot(x=found_rt.I_t_mean.index + offset, y=y, ax=ax, label='(Offset) found $\widehat{I_t}$')

# ci = found_rt.I_t_std * 1.96
ax.fill_between(found_rt.I_t_mean.index + offset,  found_rt.I_t_low, found_rt.I_t_high, color='b', alpha=.1)

# ax.set_ylim(0, 1e5)
ax.set_xlim(18, 115)
ax.set_ylabel('$\widehat{I_t}$')
ax.set_xlabel('Days')
ax.set_yscale('log')
```

```python
real_rt = pd.read_csv('synthetic_data/new_testing/no_var_bd.csv')
found_rt_with_nt = pd.read_csv('rt/synthetic/no_var_bd_60_nt.csv')
found_rt_w_no_nt = pd.read_csv('rt/synthetic/no_var_bd_60.csv')

offset = real_rt[real_rt >= 25].index[0] - 9

ax = sns.lineplot(x=real_rt.R_t.index, y=real_rt.R_t, label='Real $R_t$')
ax = sns.lineplot(x=found_rt_with_nt.R_t_mean.index - offset, 
                  y=found_rt_with_nt.R_t_mean, ax=ax, 
                  label='Found $\widehat{R_t}$ given $N_t$',
                  color='g', alpha=0.5)
ax.fill_between(found_rt_with_nt.R_t_mean.index - offset,  found_rt_with_nt.R_t_low, found_rt_with_nt.R_t_high, color='g', alpha=.1)

ax = sns.lineplot(x=found_rt_w_no_nt.R_t_mean.index - offset, 
                  y=found_rt_w_no_nt.R_t_mean, ax=ax, 
                  label='Found $\widehat{R_t}$',
                  color='r', alpha=0.5)
ax.fill_between(found_rt_w_no_nt.R_t_mean.index - offset,  found_rt_w_no_nt.R_t_low, found_rt_w_no_nt.R_t_high, color='r', alpha=.1)

ax.set_ylim(0, 4)
ax.set_xlim(0, 80)
```

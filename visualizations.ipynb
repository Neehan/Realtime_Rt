```python
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import numpy as np

sns.set()

%matplotlib inline
```

```python
found_rt = pd.read_csv('rt/synthetic/bd.csv')
real_rt = pd.read_csv('synthetic_data/bd.csv')
```

```python
scale_const = 8/7
offset = 25 - 5.2

ax = sns.lineplot(x=real_rt.R_t.index, y=real_rt.R_t, label='Real Rt')
y = found_rt.R_t_mean* scale_const
ax = sns.lineplot(x=found_rt.R_t_mean.index + offset, y=y, ax=ax, label='(Scaled, offset) found Rt')

ci = found_rt.R_t_std * 1.96
ax.fill_between(found_rt.R_t_mean.index + offset, (y-ci), (y+ci), color='b', alpha=.1)

ax.set_ylim(0.5, 3)
ax.set_xlim(18, 115)
```

```python
ax = sns.lineplot(x=real_rt.I_t.index, y=real_rt.I_t, label='Real It')
y = found_rt.I_t_mean
ax = sns.lineplot(x=found_rt.I_t_mean.index + offset, y=y, ax=ax, label='(Offset) found It')

ci = found_rt.I_t_std * 1.96
ax.fill_between(found_rt.I_t_mean.index + offset, (y-ci), (y+ci), color='b', alpha=.1)

ax.set_ylim(0, 1e5)
ax.set_xlim(18, 115)
```

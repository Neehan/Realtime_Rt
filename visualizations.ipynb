```python
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import numpy as np

sns.set()

%matplotlib inline
```

```python
found_rt = pd.read_csv('rt/synthetic/bd.csv')
real_rt = pd.read_csv('synthetic_data/bd.csv')
```

```python
ax = sns.lineplot(x=real_rt.R_t.index, y=real_rt.R_t)
ax = sns.lineplot(x=found_rt.R_t_mean.index + 25, y=found_rt.R_t_mean, ax=ax)

y = found_rt.R_t_mean
ci = found_rt.R_t_std * 1.96
ax.fill_between(found_rt.R_t_mean.index + 25, (y-ci), (y+ci), color='b', alpha=.1)

ax.set_ylim(0.5, 3)
ax.set_xlim(25, 105)
```

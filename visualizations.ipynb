```python
real_rt = pd.read_csv('synthetic_data/new_testing/constant_downward_bd.csv')
found_rt_with_nt = pd.read_csv('rt/synthetic/constant_downward_bd_60_nt.csv')
found_rt_w_no_nt = pd.read_csv('rt/synthetic/constant_downward_bd_60.csv')

offset = real_rt[real_rt >= 25].index[0] - 9

ax = sns.lineplot(x=real_rt.R_t.index, y=real_rt.R_t, label='Real $R_t$')
ax = sns.lineplot(x=found_rt_with_nt.R_t_mean.index - offset, 
                  y=found_rt_with_nt.R_t_mean, ax=ax, 
                  label='Found $\widehat{R_t}$ given $N_t$',
                  color='g', alpha=0.5)
ax.fill_between(found_rt_with_nt.R_t_mean.index - offset,  found_rt_with_nt.R_t_low, found_rt_with_nt.R_t_high, color='g', alpha=.1)

ax = sns.lineplot(x=found_rt_w_no_nt.R_t_mean.index - offset, 
                  y=found_rt_w_no_nt.R_t_mean, ax=ax, 
                  label='Found $\widehat{R_t}$',
                  color='r', alpha=0.5)
ax.fill_between(found_rt_w_no_nt.R_t_mean.index - offset,  found_rt_w_no_nt.R_t_low, found_rt_w_no_nt.R_t_high, color='r', alpha=.1)

ax.set_ylim(0, 4)
ax.set_xlim(0, 80)
```

```python
def sample_Beta_Binomial(a, b, n, size=100000):
    p = np.random.beta(a, b, size=size)
    r = np.random.binomial(n, p)
    return r

window = 175

positives, total_tests = real_rt.P_t.iloc[-window:], real_rt.T_t[-window:]
negatives = total_tests - positives
N_t = real_rt.N_t[-window:]

R_t = []
R_t_low = []
R_t_high = []

alpha, beta = positives.iloc[0], negatives.iloc[0]
N = N_t.iloc[0]
I_t_samples = sample_Beta_Binomial(alpha, beta, N)

serial_interval = 9.0
gamma = 1/serial_interval

for i in range(1, window):
    alpha_new, beta_new = alpha + positives.iloc[i], beta + negatives.iloc[i]
    N_new = N_t.iloc[i]
    I_t_1_samples = sample_Beta_Binomial(alpha_new, beta_new, N_new)
    
    # Now, generate the R_t estimate.
    R_t_samples = ((np.log(I_t_1_samples) - np.log(I_t_samples)) / gamma) + 1.
    R_t.append(np.mean(R_t_samples))
    low, high = R_t[-1] - 1.96 * np.std(R_t_samples), R_t[-1] + 1.96 * np.std(R_t_samples) 
    R_t_low.append(low)
    R_t_high.append(high)

    alpha, beta, N = alpha_new, beta_new, N_new
    I_t_samples = I_t_1_samples
    
```

```python
fig = plt.figure(figsize=(10,6))
ax = sns.lineplot(x=range(len(R_t)), 
                  y=R_t, 
                  label='$\widehat{R_t}$ (our method)',
                  color='r', alpha=0.5)
ax.fill_between(range(len(R_t)), R_t_low, R_t_high, color='r', alpha=.1)

from scipy.signal import savgol_filter
positives_growth_rate = ((np.log(1+positives.iloc[1:].to_numpy()) - np.log(1+positives.iloc[:-1].to_numpy())) / gamma) + 1
positives_growth_rate = np.clip(positives_growth_rate, 0, 12)
smoothed_positives = savgol_filter(positives_growth_rate, window_length=7, polyorder=2)
ax = sns.lineplot(x=range(len(R_t)), 
                  y=smoothed_positives, ax=ax,
                  label='$\widehat{R_t}$ using positive tests only',
                  color='g', alpha=0.5)

# Now, compare with the case growth rate
cases = real_rt.cases[-window:]
case_growth_rate = ((np.log(cases.iloc[1:].to_numpy()) - np.log(cases.iloc[:-1].to_numpy())) / gamma) + 1
ax = sns.lineplot(x=range(len(R_t)), 
                  y=case_growth_rate, ax=ax,
                  label='True case growth rate',
                  color='b', alpha=0.5)

# Now, compare with the case growth rate
ax.set_ylim(0, 4)
ax.set_xlabel('Days')
ax.set_ylabel('$R_t$')
ax.set_title('Comparison between $R_t$ estimation methods')

plt.savefig('R_t_comparison.png', dpi=400)
```
